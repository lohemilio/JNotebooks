# Selección de muestra representativa por escuela

## Importación de librerías

```python
import pandas as pd
```

## Importación de datos

```python
profesionalData = pd.read_excel(r'DataSets/Ensayos_profe_v2.xlsx')
```

```python
profesionalData.head()
```

## División por escuela

```python
profesionalData.Escuela.unique()
```

###                        Ingeniería y ciencias

```python
#TODOS LOS REGISTROS
profesionalData[profesionalData.Escuela == 'Ingeniería y Ciencias']
```

####      Muestra de ingeniería y ciencias: 339 registros

```python
#RANDOM SELECTION
ingenieriaData = profesionalData[profesionalData.Escuela == 'Ingeniería y Ciencias']
muestraIngenieria = ingenieriaData.sample(n=339)
muestraIngenieria.head()
```
